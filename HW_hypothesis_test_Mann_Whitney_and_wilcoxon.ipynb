#First Case: Playcatch

Playcatch is a streaming platform that recently implemented a major update, introducing new features and a redesigned interface. The development team is interested in evaluating whether this update has led to a significant improvement in user ratings compared to a competing platform.









```python
import pandas as pd
import numpy as np
from statistics import mean
```

```python
df_avaliacoes = pd.DataFrame({
    'playcatch': [4, 5, 3, 4, 5, 4, 4, 3, 5, 4, 5, 3, 4],
    'concorrente': [3, 4, 3, 3, 2, 4, 3, 2, 3, 3, 4, 2, 3]
})
df_avaliacoes
```

```python
mean(df_avaliacoes['playcatch'])
```

```python
mean(df_avaliacoes['concorrente'])
```

```python
#hypothesis
# H0: There is no significant difference in the average ratings between Playcatch and the competitor.
# H1: There is a significant difference in the average ratings between Playcatch and the competitor.
```

```python
from scipy.stats  import mannwhitneyu
```

```python
stats, p_valor = mannwhitneyu(df_avaliacoes['playcatch'], df_avaliacoes['concorrente'], alternative = "two-sided")
```

```python
print("Statistic:", stats)
print("P-valor:", p_valor)
```

```python
# We can reject the null hipothesis
```

```python
nivel_significancia = 0.05

if p_valor < nivel_significancia:
  conclusao = "Reject null hypothesis"
else:
  conclusao = "Not Reject null hypothesis"

conclusao
```

#Second Case: Playcatch

In this new case, Playcatch introduced a new interactive feature on its platform called WatchTogether, which allows users to watch content simultaneously and share experiences in real time.

For the study, the company collected user ratings (on a scale from 1 to 5 stars) before and after the introduction of the new feature. Below are the collected scores:









```python
df_wt = pd.DataFrame({
    'Antes_WatchTogether': [3, 4, 3, 4, 2, 4, 3, 2, 3, 3, 4, 2, 3],
    'Depois_WatchTogether': [4, 5, 3, 4, 5, 4, 4, 3, 5, 4, 5, 3, 4]
})

df_wt
```

```python
#diferenças de percepção
resultado_subtracao = [b - a for a, b in zip(df_wt['Antes_WatchTogether'],df_wt['Depois_WatchTogether'])]
resultado_subtracao
```

```python
np.median(resultado_subtracao)
```

```python
# Hypothesis
# H0 = Median of diferences betwens pairs is zero
# H1 = Median of diferences betwens pairs isn't zero
```

```python
from scipy.stats import wilcoxon
stats, p_valor = wilcoxon(resultado_subtracao)
```

```python
nivel_significancia = 0.05

if p_valor < nivel_significancia:
  conclusao = "Reject null hypothesis"
else:
  conclusao = "Not Reject null hypothesis"

conclusao
```
